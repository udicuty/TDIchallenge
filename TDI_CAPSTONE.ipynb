```python
comb=pd.concat([cave,hammock])
comb.columns = map(lambda x: x.replace(' ','_'), comb.columns)
comb
```

```python
#Replacing dates
for i in range(len(comb.columns)):
    if 'Date' in comb.columns[i]:
        a=comb.columns[i]
        comb[a]=pd.to_datetime(comb[a])

comb["Processing"]=comb.Closed_Date-comb.Created_Date
comb.index=comb.Unique_Key
comb=comb.sort_values("Created_Date",ascending=False)
```

```python
def convert_to_days(timed):
    return(timed.total_seconds()/3600/24)

q=[convert_to_days(delta[1]) for delta in comb.Processing.iteritems()]

q=pd.Series(q)
av=q.median()
q[np.isnan(q)]=av
q[q<0]=av
q[q>50]=av
#del q[q<0]
#np.mean(q)
sns.distplot(list(q),bins=10)
#q.notnan
```
